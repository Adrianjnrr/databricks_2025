```python
from pyspark.sql.functions import col
#user_name = dbutils.notebook.entry_point.getDbutils().notebook().getContext().userName().get()
files = ['flights', 'airports', 'airlines']

for f in files:
    file_location = f'/Volumes/workspace/aero/{f}.csv'
    file_type = "csv"

    # CSV options
    infer_schema = "true"
    first_row_is_header = "true"
    delimiter = ","

    # The applied options are for CSV files. For other file types, these will be ignored.
    df = spark.read.format(file_type) \
    .option("inferSchema", infer_schema) \
    .option("header", first_row_is_header) \
    .option("sep", delimiter) \
    .load(file_location)
    df.write.saveAsTable(f'aero.{f}')
```

Find the top 20 shortest flight by Airtime, show where such regular flights are relized
Show their full name of the airports from (Origin) and to (Dest)
Add also Latitude and Longitude and visualize there locations of the airport on thr map

```python
%sql
create temporary View shortest_flight as select f.Origin, 
f.OriginCityName,
ao.Latitude as OriginLat,
ao.Longitude as OriginLon,
f.Dest,
f.DestCityName,
ad.Latitude as DestLat,
ad.Longitude as DestLong,
f.AirTime
from aero.flights f
join aero.airports ao on f.Origin = ao.IATA_CODE
join aero.airports ad on f.Dest = ad.IATA_CODE
where f.AirTime is not null
order by f.AirTime asc
limit 20
```

```python
%sql
select Origin as Airport, OriginLat as Latitude, OriginLon as Longitude from shortest_flight
union 
select Dest as Airport, DestLat as Latitude, DestLong as Longitude from shortest_flight;
    

    

    
```
